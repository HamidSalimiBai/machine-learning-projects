```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
x = df5.drop('Outcome', axis = 1).values
y = df5.Outcome.values.reshape(-1 ,1)
```

```python
x
```

```python
x_train,x_test, y_train, y_test = train_test_split(x, y, test_size = 0.35, random_state = 0)
```

```python
k = 5
clf3 = KNeighborsClassifier(5)
clf3.fit(x_train, y_train.ravel())
y_pred = clf3.predict(x_test)
```

```python
print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
```

### to understand the best value of k we need to right a loop

```python
k = 20
ACC = np.zeros((k))

for i in range(1, k+1 ):
    clf3 = KNeighborsClassifier(n_neighbors = i)
    clf3.fit(x_train, y_train.ravel())
    y_pred = clf3.predict(x_test)
    ACC[i -1] = metrics.accuracy_score(y_test, y_pred)
ACC
```

```python
print(np.max(ACC))
print(np.min(ACC))
```

```python
training_acc = []
testing_acc = []

neighbors_setting  = range(1 ,50)

for n_neighbors in neighbors_setting:
    knn = KNeighborsClassifier(n_neighbors = n_neighbors)
    knn.fit(x_train, y_train.ravel())
    training_acc.append(knn.score(x_train, y_train))
    testing_acc.append(knn.score(x_test, y_test))
    
plt.plot(neighbors_setting, training_acc, label = 'Accuracy of the training_acc')
plt.plot(neighbors_setting, testing_acc, label = 'Accuracy of the testing_acc')
plt.xlabel('number of neighbors')
plt.ylabel('Accuracy')
plt.grid()
plt.legend
```

### improve model with Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'n_neighbors': range(1, 50)}
grid_knn = GridSearchCV(estimator = knn,
                       param_grid = parameters,
                       scoring = 'accuracy',
                       cv = 5,
                       verbose = 1,
                       n_jobs = -1)
grid_knn.fit(x_train, y_train.ravel())
```
