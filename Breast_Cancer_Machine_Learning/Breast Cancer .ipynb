# Prepare for Machine learning

```python
data['Marital_Status'].replace({'Married':2, 'Divorced':3, 'Single ':4, 'Widowed':5, 'Separated':6}, inplace = True )
data['T_Stage'].replace({'T1':2, 'T2':3, 'T3':4, 'T4':5}, inplace = True)
data['N_Stage'].replace({'N1':2, 'N2':3, 'N3':4},inplace = True )
data['6th_Stage'].replace({'IIA':2, 'IIIA':3, 'IIIC':4, 'IIB':5, 'IIIB':6}, inplace = True)
data['differentiate'].replace({'Poorly differentiated':2, 'Moderately differentiated':3, 'Well differentiated':4, 'Undifferentiated':5}, inplace = True)
data['A_Stage'].replace({'Regional':2, 'Distant':3}, inplace = True)
data['Estrogen_Status'].replace({'Positive':2, 'Negative':3}, inplace = True)
data['Progesterone_Status'].replace({'Positive':2, 'Negative':3}, inplace = True)
data['Status'].replace({'Alive':1, 'Dead':0}, inplace = True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import confusion_matrix, classification_report
```

```python
x = data.drop('Status', axis = 1)
y = data.Status.values.reshape(-1, 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.15, random_state = 0)
```

```python
model = LogisticRegression(solver = 'liblinear')
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```
